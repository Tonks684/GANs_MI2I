## ACT4 Confidence Map



This notebook is developed to generate a pixel wise confidence map for each Pix2PixHD GAN prediction. We are able to generate a confidence map by exploiting the non-deterministic nature of the method. Using the /source_code/pix2pixHD/variationa_inference.py file for a single image we generate multiple predictions each with **Dropout** applied. </br> 

Dropout introduces randomness into the predictions and we show a very **strong positive correlation between variance and error**. This leads to the hyposthsis that high variance is a proxy for high error which we quantify as **low confidence**. While low variance is a proxy for low error which we quantify as **high confidence**. 


This notebook will generate the image prediction stack along with the two-dimensional confidence map

```python
outputs_stacks = [i for i in Path(variational_inf_path).glob('**/*1.tiff')]
print(len(outputs_stacks))
print(outputs_stacks[:3])
```

```python
def normalize(arr):
    """
    8bit normalisation.

    :param arr:
    :return:
    """
    scale_arr = (arr - 85.0) / (12573.0 - 85.0)
    scale_arr = scale_arr * 255
    scale_arr = scale_arr.astype(np.uint8)
    return scale_arr
```

```python
for output in outputs_stacks:
    print(output)
    folder = str(output).split('/')[-1].split('_')[1][:-8]
    Path(f'{variational_inf_path}/{folder}/16bit_images/').mkdir(parents=True,exist_ok=True)
    Path(f'{variational_inf_path}/{folder}/8bit_images_mask/').mkdir(parents=True,exist_ok=True)
    
    tmp_stack = imread(str(output))
    print(tmp_stack.shape)
    for i in range(1000):
        img = tmp_stack[i, :, :]
        imsave(f'{variational_inf_path}/{folder}/16bit_images/3CG0022780_{folder}_synthesized_{i}.tif',img.astype(np.float32))
        img_8bit = normalize(img)
        imsave(f'{variational_inf_path}/{folder}/8bit_images_mask/3CG0022780_{folder}_synthesized_{i}.tif',img.astype(np.uint8))
```
