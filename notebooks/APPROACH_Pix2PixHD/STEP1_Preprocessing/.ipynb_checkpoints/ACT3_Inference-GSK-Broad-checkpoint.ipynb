### Define general hyper-parameters

 

```python
gsk_broad_path = os.path.join('/hpc/scratch/rdip1/smt29021/XAI/ELN128360/BFcellhealth_20230322_113718/AssayPlate_PerkinElmer_CellCarrierUltra/')
gsk_broad = sorted([p for p in Path(f'{gsk_broad_path}/').glob(f'**/*.tif')])
```

```python
folders = set([str(i).split('/')[-1].split('_')[-2] for i in gsk_broad])

element_count = Counter([str(i).split('/')[-1].split('_')[-2] for i in gsk_broad])
grouped_by_count ={}
for element, count in element_count.items():
    if count not in grouped_by_count:
        grouped_by_count[count] = [element]
    else:
        grouped_by_count[count].append(element)
print('Group by count:', grouped_by_count)
```

### Define model hyper-parameters

See ../../source_code/pix2pixHD for detailed breakdown of all hyper-parameters.

```python
command_list = []
arg_dict = {}
    
# Align with train
arg_dict['--checkpoints_dir'] = model_dir
arg_dict['--dataroot'] = input_folder
arg_dict['--data_type'] = '16'
arg_dict['--label_nc'] = '0'
arg_dict['--input_nc'] = '1'
arg_dict['--output_nc'] = '1'
arg_dict['--name'] = model_name
arg_dict['--no_instance'] = ''
arg_dict['--resize_or_crop'] = 'none'
arg_dict['--gpu_ids'] = '0,1,2,3'
arg_dict['--loadSize'] = '1000'
arg_dict['--norm'] = 'instance'
arg_dict['--output_reshape'] = '1000'  

# Inference only
arg_dict['--how_many'] = total_test_set_images
arg_dict['--results_dir'] = os.path.join(output_dir,output_folder_name)

create_directory(os.path.join(output_dir,output_folder_name,model_name,"test_"+epoch_no))
arg_dict['--which_epoch'] = epoch_no 
    
command = test_pix2pixHD(py_file,arg_dict)
command_list.append(command)
    
print("Number of Jobs {}".format(len(command_list)))
print(command_list[:1])
      
```

### Generate slurm command and schedule

```python
command = test_pix2pixHD(
    py_file,
    arg_dict
)

command_list = []
command_list.append(command)
print(command_list)
```

```python
### Job Settings ###
job_name = '{}'.format(model_name)

node_setting = ''
node_setting = node_setting+' --job-name={}'.format(job_name)
node_setting = node_setting+' --time=10-00:00'
node_setting = node_setting+' --nodes=1'
# node_setting = node_setting+' --dependency=afterok:{}'.format('43523352')

node_setting = node_setting+' --partition=gpu'
node_setting = node_setting+' --gres=gpu:gtx1080ti:4'
node_setting = node_setting+' --ntasks-per-node=1'
node_setting = node_setting+' --output=./slurm_outs/"slurm-%A_%a.out"'
node_setting = node_setting[1:]

os.makedirs('./slurm_outs', exist_ok=True)

submit_array(root_dir, command_list, node_setting, job_name,repo_path, conda_path)
```

## Post inference image processing

### Save as 8bit for CellPose after Mask creation

```python
from tqdm import tqdm
from tifffile import imread, imsave
import numpy as np
for f in tqdm(folders):
    path = os.path.join(output_dir,output_folder_name,model_name,"test_"+epoch_no,f)
    imgs = sorted([p for p in Path(f'{path}/').glob(f'**/*_virtualstain.tif')])
#     print(len(imgs))
    for i in imgs:
        x = imread(str(i))
        scale_arr = (x - x.min()) / (x.max() - x.min())
        scale_arr = scale_arr * 255
        scale_arr = scale_arr.astype(np.uint8)
        imsave(os.path.join(path,i), scale_arr.astype(np.uint8),imagej=True)
```
